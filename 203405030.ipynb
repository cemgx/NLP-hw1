```python
word2idx = {}
idx2word = {}
tur_word2idx = {}
tur_idx2word = {}
```

```python
i = 0
for cell in oyuncular:
    words = cell.split()
    for word in words:
        if word in word2idx:
            continue
        word2idx[word] = i
        idx2word[i] = word
        i += 1
j = 0
for cell in tur:
    words = cell.split()
    for word in words:
        if word in tur_word2idx:
            continue
        tur_word2idx[word] = j
        tur_idx2word[j] = word
        j += 1
```

```python
def cos_sim(v1, v2):
    return np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
```

```python
def film_2_str(row_index):
    film = df.iloc[row_index]
    adi = film['film adi']
    turu = film['tur']
    ozet = film['ozet ve detaylar']
    oyuncular = film['oyuncular']
    if len(ozet) > 150:
        ozet = ozet[:150] + "..."

    return f"Film adı: {adi} \nTürü: {turu} \nOyuncular: {oyuncular} \nÖzeti: {ozet}"
```

```python
def most_similar_films(film_adi, benzer_film_sayisi):
    film_indexi = df[df["film adi"] == film_adi].index[0]
    print("Aranan Film:")
    print(film_2_str(film_indexi))
    print()
    print("* " * 30)
    print()

    benzerlikler = []
    
    tur_benzerlikler = []
    

    sorgu_film = filmler_tf_idf[film_indexi]
    
    tur_sorgu_film = tur_filmler_tf_idf[film_indexi]

    for vec in filmler_tf_idf:
        benzerlik = cos_sim(sorgu_film, vec)
        benzerlikler.append(benzerlik)
        
    for vec in tur_filmler_tf_idf:
        tur_benzerlik = cos_sim(tur_sorgu_film, vec)
        tur_benzerlikler.append(tur_benzerlik)        
    
    print(np.array(tur_benzerlikler).shape)
    
    total_benzerlik = np.add(np.array(tur_benzerlikler)*0.4,np.array(benzerlikler)*0.6)
    
    en_benzer_film_indexleri = (- np.array(total_benzerlik)).argsort()[:benzer_film_sayisi + 1]
    print("Benzer Filmler:\n")
    for i, film_idx in enumerate(en_benzer_film_indexleri[1:]):
        print(str(i+1) + ") " + film_2_str(film_idx))
        print()
        print("- " * 20)
        print()
    
```
